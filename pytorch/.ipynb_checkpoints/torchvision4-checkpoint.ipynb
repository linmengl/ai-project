```python
from PIL import Image
from torchvision import transforms

center_crop = transforms.CenterCrop((60,70))
random_crop = transforms.RandomResizedCrop((60,70))
five_corp = transforms.FiveCrop((60,70))

h_flip_oper = transforms.RandomHorizontalFlip(1)
v_flip_opr = transforms.RandomVerticalFlip(1)


image = Image.open("aa.jpg")

display(image)

display(center_crop(image))

print("-------1--------")
display(random_crop(image))
print("-------2--------")
five_corps = five_corp(image)
num = 3
for five in five_corps:
    print("--------"+str(num)+"-------")
    display(five)
    num += 1

display(h_flip_oper(image))
display(v_flip_opr(image))
```
